```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1)
```

```python
data.head()
```

```python
data['Geography'].unique()
```

```python
data = pd.get_dummies(data,drop_first=True)
```

```python
data.head()
```

```python
data['Exited'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(data=data, x='Exited', palette='Set1')
plt.xlabel('Exited')
plt.ylabel('Count')
plt.title('Distribution of Exited')
plt.savefig('visuals/Distribution of Exited.png')
plt.show()
```

```python
X = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
sns.countplot(x=y_res, palette='magma')
plt.xlabel('Exited')
plt.ylabel('Count')
plt.title('Exited Distribution Check - After SMOTE')
plt.savefig('visuals/Class Distribution After SMOTE.png')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

### 12. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```
